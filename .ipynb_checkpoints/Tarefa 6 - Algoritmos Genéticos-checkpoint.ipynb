# Algoritmos Genéticos - Aritana Noara

## Definições

### Problema

Maximizar função Alpine2: prod(sqrt(x) .* sin(x), 2)

Domínio da função: [0,10]

Função Objetivo: prod(sqrt(x) .* sin(x), 2)


### Indivíduo

O indivíduo possui uma representação cromossomica em (Xi, Xj), onde i,j são as dimensões da função alpine2.

Xi: representação da dimensão i de um cromossomo.

Xj:representação da dimensão j de um cromossomo.

Xi, Xj: Podem assumir valores no intervalo [0,1] com precisão de 3 casas decimais.

A função objetivo em função desses valores será:

Alpine2 = sqrt(Xi) * sin(Xi)* sqrt(Xj) * sin(Xj)

### Populacão

Aleatória.


### Gerações

Aleatória

### Função de Fitness

Iremos utilizar o método de Ranking Linear:
FuncaoFitness =  min + ( max - min) * (n-1)/(n-1)

onde:
i: posição do vetor de indivíduos ordenado.
n: tamanho da população.
min: valor designado ao indivíduo mais apto.

### Normalização Linear

Quanto maior a constante de incremento (max - min) / (pop_zize- 1), maior a pressão seletiva sobre os melhores (privilégio aos melhores).
O objetivo será encontrar uma taxa para evitar convergência rápida. Iremos utilizar taxa = 10 inicialmente e tentar otimizar conforme resultados apresentados.

### CrossOver

Utilizaremos o crossover aritmético que consiste em uma combinação linear de dois cromossomos:

O1 =  aP1 + (1-a)P2

O2 =  aP2 + (1-a)P2

Sendo:

P1: pai1

P2: pai2

O1: filho 1

O2: filho 2

a = numero aleatório no intervalo [0,1]





```python
#Inicialização de variáveis

tam_pop = 70 #tamanho da populacao
geracoes = 50 #numero de geracoes.
Min=1 #ranking linear
Max=30 #ranking linear
N=tam_pop #ranking linear
```

```python
#Função objetivo
import math

def alpine2(xi,xj):
    alpine = math.sqrt(xi)*math.sin(xi)*math.sqrt(xj)*math.sin(xj)
    return alpine

teste = alpine2(5.689,6.281)
print(teste)
```

```python
#Individuo
import random

class Individuo:
    def __init__(self, xi,xj,fx,aptidao=-999999999):#xi,xj e a alpine2(xi,xj)
        self.xi = xi
        self.xj = xj
        self.fx = fx
        self.aptidao = aptidao
    
    def toString(self):
        array=[]
        array.append(self.xi)
        array.append(self.xj)
        array.append(self.fx)
        array.append(self.aptidao)
        return array

def gerar_individuo():  
    xi = random.uniform(0, 10)#entre 0 e 10
    xj = random.uniform(0, 10)#entre 0 e 10   
    fx = alpine2(xi, xj)
    individuo =Individuo(xi,xj,fx)
    return individuo


teste = gerar_individuo()
print(teste.xi)
print(teste.xj)
print(teste.fx)
print(teste.toString())
```

```python
#Mostrar indivíduos
def mostrarIndividuos(populacao):
    for ind in populacao:
        print('Indivíduo(xi,xj):(%f, %f), alpine2:%f, fitness: %f' \
          %(round(ind.xi,3), round(ind.xj,3),round(ind.fx,3),round(ind.aptidao,3)))



#teste =  mostrarIndividuos(populacao)#com fitness ainda não definida
```

```python
#Populacao
        
def gerar_populacao (tam_pop):
    populacao=[]
    for i in range(tam_pop):     
        populacao.append(gerar_individuo())        
    return populacao

populacao=gerar_populacao(tam_pop)#gerando populacao

#teste

mostrarIndividuos(populacao)

print(len(populacao))
```

```python
#Função fitness: recebe uma populacao ordenada e retorna a populacao com informacoes sobre aptidao de cada indiv.


def valorDaFuncaoObjetivo(valores):#ordenacao
    valor =  -valores.fx    
    return valor


def ordenarIndividuosPorFuncaoObjetivo(populacao):#ordenacao
    pop_ord=[]
    for ind in populacao:        
        pop_ord = sorted(populacao,key=valorDaFuncaoObjetivo)
    return pop_ord

def fitnessRankingLinear(populacaoOrdenada,Min, Max, N):    
    populacao =[]
    for k in range(len(populacaoOrdenada)):
        i = k+1 #primeira posicao = 1
        fx = Min + (Max - Min)*(N - i ) / (N - 1)        
        populacaoOrdenada[k].aptidao=fx
        populacao.append(populacaoOrdenada[k])
    return populacao


#teste
populacaoOrdenada = ordenarIndividuosPorFuncaoObjetivo(populacao)
print ("ordenada")
mostrarIndividuos(populacaoOrdenada)

print ("fitness")
populacao = fitnessRankingLinear(populacaoOrdenada,Min, Max, N)
mostrarIndividuos(populacao)
 


```

```python
#Selecionar os Pais por método da roleta

def somarAptidaoIndividuos(populacao, tam_pop):
    total = 0
    for ind in populacao: 
        total += ind.aptidao
    return total

def numeroAleatorioEmIntervaloFechado(At):#retorna numero entre 0 e AT
    valor= round((random.uniform(0, At)),3)
    return valor

def vetorAptidoesAcumuladas(populacao, tam_pop):
    vetor=[]
    aptidaoAnterior=0  

    for i in range(len(populacao)-1,-1,-1): #start, end, step  
        info={
        "indice":None,
        "aptidaoAcum":None,        
         }
        ind = populacao[i]
        aptidao =  ind.aptidao + aptidaoAnterior
        info["indice"]=i
        info["aptidaoAcum"]=round(aptidao,3)
        vetor.append(info)
        aptidaoAnterior = round(aptidao,3)   
    return vetor
    

def vetorNumerosAleatorios(tam_pop,At):#intervalos para numero aleatorio
    numerosAleatorios = []
    for i in range(tam_pop): 
        numeroAleatorio=numeroAleatorioEmIntervaloFechado(At)
        numerosAleatorios.append(numeroAleatorio)
    return numerosAleatorios

def roleta(populacao,tam_pop):
    indicesPaisSelecionados=[]
    At = somarAptidaoIndividuos(populacao, tam_pop)
    numerosAleatorios = vetorNumerosAleatorios(tam_pop,At)
    aptidoesAcumuladas = vetorAptidoesAcumuladas(populacao, tam_pop)  
    print(At)
    print(numerosAleatorios)
    print()
    print(aptidoesAcumuladas)
    for rand in numerosAleatorios:
        for aptidao in aptidoesAcumuladas:
            if aptidao['aptidaoAcum'] >= rand:
                #print('\nrand: %f aptidao %f'%(rand,aptidao['aptidaoAcum']))
                indicesPaisSelecionados.append(aptidao['indice'])#indice da populacao ordenada por fitness (passo anterior)
                break
    return indicesPaisSelecionados      

#teste
indicePaisSelecionados = roleta(populacao,tam_pop)
print('Indice dos Pais Selecionados')
print(indicePaisSelecionados)
#print('aptidao anterior:%f'%(aptidaoAnterior))
```

```python
#Pais Selecionados
def paisSelecionados(populacao):
    indicePaisSelecionados = roleta(populacao,tam_pop)
    novaPopulacao=[]
    for indice in indicePaisSelecionados: 
        novaPopulacao.append(populacao[indice])
    return novaPopulacao

        
#Teste
pais = paisSelecionados(populacao)

print("Pais selecionados")
mostrarIndividuos(pais)
```

```python
#Gerar_filhos - Crossover aritmético
def gerarUmFilho(a,p1,p2):   
    b = 1-a    
    p1xi = a*p1.xi
    p1xj= a*p1.xj
    
    p2xi = b*p2.xi
    p2xj= b*p2.xj
    
    xi = p1xi + p2xi
    xj = p1xj + p2xj
    fx = alpine2(xi, xj)   
    filho =Individuo(round(xi,3),round(xj,3),round(fx,3))
    return filho

def crossover(pais, tam_pop, taxa_crossover=0.7):#crossover minimo de 70%
    nova_pop=[] 
    for i in range(tam_pop//2):#divisao inteira, seleciono dois pais e gero dois filhos
        pai1 =  random.choice(pais)#pai1
        pai2= random.choice(pais)
        if random.random() < taxa_crossover:#crossover de um ponto
            a = (random.uniform(0, 1))   #gera um novo numero alfa no int [0,1]              
            filho1= gerarUmFilho(a,pai1,pai2)
            filho2= gerarUmFilho(a,pai2,pai1)
            nova_pop.append(filho1)
            nova_pop.append(filho2)            
        else: # se nao ocorrer o crossover, os pais continuam na populacao
            nova_pop.append(pai1)
            nova_pop.append(pai2) 
    return nova_pop  



    
#teste         
novaPopulacao = crossover(pais, tam_pop, taxa_crossover=0.7)
populacaoParaSerMutada = novaPopulacao
mostrarIndividuos(novaPopulacao)
print('Populacao para ser mutada')
mostrarIndividuos(populacaoParaSerMutada)

#precisa ordenar e fazer fitness
print()
populacaoOrdenada = ordenarIndividuosPorFuncaoObjetivo(populacaoParaSerMutada)
populacao = fitnessRankingLinear(populacaoOrdenada,Min, Max, N)
print("Populacao para ser mutada apos funcao fitness")
mostrarIndividuos(populacao)
populacaoParaSerMutada = populacao
 
       
```

```python
#Mutacao

def mutarPopulacao(populacao,taxa_mutacao = 0.1):
    nova_pop=[]
    limite = [0,1]
    for ind in populacao: 
        if random.random()<taxa_mutacao:  
            change =  random.choice(limite)
            if change == 0:
                delta = -0.01
            else:
                delta = 0.01  
            ind.xi = ind.xi + delta
            ind.xj = ind.xj+ delta
            ind.fx = alpine2(ind.xi,ind.xj)
            nova_pop.append(ind)
        else: #nao mutou
            nova_pop.append(ind)
    return nova_pop    

#teste
print("Nova Populacao")
novaPopulacao = mutarPopulacao(populacaoParaSerMutada)
mostrarIndividuos(novaPopulacao)

```

```python
def maximumFitness(pop,tam_pop):
    melhor_avaliacao = pop[0].aptidao #o primeiro.
    indice_melhor =0    
    for i in range(1, len(pop)):
        avaliacao = pop[i].aptidao
        if avaliacao > melhor_avaliacao:
            melhor_avaliacao = avaliacao
            indice_melhor = i
    return pop[indice_melhor].aptidao


def maximumAlpine2(pop,tam_pop):
    melhor_alpine = pop[0].fx #o primeiro.
    indice_melhor =0    
    for i in range(1,len(pop)):
        avaliacao = pop[i].fx
        if avaliacao > melhor_alpine:
            melhor_alpine = avaliacao
            indice_melhor = i
    return(pop[indice_melhor].fx)

#teste
maximumF = maximumFitness(novaPopulacao, tam_pop)
print(maximumF)

#teste
maximumAlpine2(novaPopulacao, tam_pop)
print(maximumAlpine2(novaPopulacao,tam_pop))
```

```python
def minimumFitness(pop, tam_pop):#menos apto
    pior_avaliacao = pop[0].aptidao #o primeiro.
    indice_pior =0    
    for i in range(1, len(pop)):
        avaliacao = pop[i].aptidao
        if avaliacao < pior_avaliacao:
            pior_avaliacao = avaliacao
            indice_pior = i
    return pop[indice_pior].aptidao

#teste
minimum = minimumFitness(pais, tam_pop)
print(minimum)
```

```python
def averageFitness(pop, tam_pop):#menos apto   
    total = 0
    for i in range(len(pop)):
        total += pop[i].aptidao
    media = total/tam_pop  
    return media

#teste
media =  averageFitness(pais, tam_pop)
print(media)
```
