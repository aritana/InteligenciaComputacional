# Matriz de adjacência
Para mapearmos as distâncias entre as cidades do mapa.:



### Executemos a matriz de adjacência abaixo:

```python
class Matriz_Adjacência:
    def __init__(self,vertices):
        self.vertices =  vertices
        self.grafo = [[-1]*self.vertices for i in range(self.vertices)]
    
    def adiciona_aresta(self, linha, coluna,valor):
        self.grafo[linha-1][coluna-1] = valor
    
    def mostra_matriz(self):
        print('A matriz de adjacência é:')
        for i in range(self.vertices):
            print(self.grafo[i])
    
    def pega_valor_celula(self, linha, coluna):
        return self.grafo[linha][coluna]

        

g = Matriz_Adjacência(5)


g.adiciona_aresta(1,1,0.0)
g.adiciona_aresta(1,2,1.0)
g.adiciona_aresta(1,3,2.2)
g.adiciona_aresta(1,4,2.0)
g.adiciona_aresta(1,5,4.1)


g.adiciona_aresta(2,1,1.0)
g.adiciona_aresta(2,2,0.0)
g.adiciona_aresta(2,3,1.4)
g.adiciona_aresta(2,4,2.2)
g.adiciona_aresta(2,5,4.0)

g.adiciona_aresta(3,1,2.2)
g.adiciona_aresta(3,2,1.4)
g.adiciona_aresta(3,3,0.0)
g.adiciona_aresta(3,4,2.2)
g.adiciona_aresta(3,5,3.2)

g.adiciona_aresta(4,1,2.0)
g.adiciona_aresta(4,2,2.2)
g.adiciona_aresta(4,3,2.2)
g.adiciona_aresta(4,4,0.0)
g.adiciona_aresta(4,5,2.2)


g.adiciona_aresta(5,1,4.1)
g.adiciona_aresta(5,2,4.0)
g.adiciona_aresta(5,3,3.2)
g.adiciona_aresta(5,4,2.2)
g.adiciona_aresta(5,5,0.0)

g.mostra_matriz()





        
```
